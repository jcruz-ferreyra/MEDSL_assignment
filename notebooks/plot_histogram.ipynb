```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load cleaned data
OUTPUT_DIR = Path.cwd() / ".." / "data/processed"
df = pd.read_csv(OUTPUT_DIR / "part1_clean.csv")

# Create distribution plot
fig, ax = plt.subplots(figsize=(10, 6))

# Filter out zero votes for better visualization
votes_nonzero = df[df['votes'] > 0]['votes']
votes_nonzero = df[df['votes'] < 100000]['votes']

ax.hist(votes_nonzero, bins=50, edgecolor='black', alpha=0.7)
ax.set_xlabel('Votes', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Distribution of Votes Across All Candidates', fontsize=14)
ax.set_yscale('log')

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'vote_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"Total candidate entries: {len(df)}")
print(f"Entries with votes > 0: {len(votes_nonzero)}")
print(f"Vote statistics:\n{votes_nonzero.describe()}")
```

```python
# Is 999999 used elsewhere?
print(f"Records with 999999 votes: {(df['votes'] == 999999).sum()}")

# What do other Cedar County precincts look like?
print("\nCedar County vote counts:")
print(df[df['county_name'] == 'CEDAR COUNTY'].groupby('precinct')['votes'].sum().sort_values(ascending=False))
```
